```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.dropna(inplace=True)
```

```python
df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
CategoryString = df["Category"]
categoryVal = df["Category"].unique()
categoryValCount = len(categoryVal)
category_dict = {}
for i in range(0,categoryValCount):
    category_dict[categoryVal[i]] = i
df["Category_c"] = df["Category"].map(category_dict).astype(int)
```

```python
def replaceMultiple(mainString, toBeReplaces, newString):
    for elem in toBeReplaces :
        if elem in mainString :
            mainString = mainString.replace(elem, newString)
    return  mainString   
```

```python
for i in range(len(df)):
    df['Size'][i]=replaceMultiple(df['Size'].reset_index().iloc[:,1][i],[','],'')
```

```python
for i in range(len(df)):
    df['Installs'][i]=replaceMultiple(df['Installs'].reset_index().iloc[:,1][i],[',','+'],'')
```

```python
def change_size(size):
    if 'M' in size:
        x = size[:-1]
        x = float(x)*1000000
        return(x)
    elif 'k' == size[-1:]:
        x = size[:-1]
        x = float(x)*1000
        return(x)
    else:
        return None

df["Size"] = df["Size"].map(change_size)

#filling Size which had NA
df.Size.fillna(method = 'ffill', inplace = True)
```

```python
df['Installs']=df['Installs'].astype(float)
```

```python
df['Installs']=df['Installs'].replace('100,000+','100000')
df['Installs']=df['Installs'].replace('1,000,000+','1000000')
df['Installs']=df['Installs'].replace('500,000+','500000')
df['Installs']=df['Installs'].replace('10,000,000+','10000000')
df['Installs']=df['Installs'].replace('50,000+','50000')
df['Installs']=df['Installs'].replace('10,000+','10000')
df['Installs']=df['Installs'].replace('100,000,000+','100000000')
df['Installs']=df['Installs'].replace('5,000,000+','5000000')
df['Installs']=df['Installs'].replace('50,000,000+','50000000')
df['Installs']=df['Installs'].replace('1,000,000,000+','1000000000')
df['Installs'].value_counts()
```

```python
df['Price']=df.Price.str.replace('$','').astype(float)
```

```python
def type_cat(price):
    if price!='0':
        return 1
    else:
        return 0

df['Type'] = df['Price'].map(type_cat)
```

```python
RatingL = df['Content Rating'].unique()
RatingDict = {}
for i in range(len(RatingL)):
    RatingDict[RatingL[i]] = i
df['Content Rating'] = df['Content Rating'].map(RatingDict).astype(int)
```

```python
df.drop(labels = ['Last Updated','Minimum Version','Latest Version','App Name'], axis = 1, inplace = True)
```

```python
def price_clean(price):
    if price == '0':
        return 0
    else:
        price = price[1:]
        price = float(price)
        return price

df['Price'] = df['Price'].map(price_clean).astype(float)
```
